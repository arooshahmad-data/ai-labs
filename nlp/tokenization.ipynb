# Tokenizing with NLTK

## Word Tokenizer

```python
!pip install nltk
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

text = "Once a big fat brown fox jumps very quickly over a lazy and hungry dog."

tokenized = word_tokenize(text)
print(tokenized)
```

## Sentence Tokenize

```python
from nltk.tokenize import sent_tokenize

text = "Hi Arush!. How are you doing?. Did you have breakfast today?."

tokenized = sent_tokenize(text)
print(tokenized)

```
